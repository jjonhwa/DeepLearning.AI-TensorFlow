# **The Hello World of Deep Learning with Neural Networks**

선형식을 이용한 매우 간단한 예제를 통해 알아보도록 하자.  
`float how_function(float x){
    float y = (2 * x) - 1;  
    return y;  
}`  

X & Y의 데이터를 이용하여 우리는 위의 식을 훈련시킬 수 있다.  
이것은 우리들이 사용했던 방식하고는 완전히 다른 Paradigm이다.

# Imports

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

# Define and Compile the Neural Network

Neural Network를 사용할 수 있는 간단한 예제를 만들어보도록 하자.  
다음은 1layer에 1neuron만이 있는 아주 간단한 예제이다.

```
model  = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
```

Neural Network를 실행하기 위해 우리는 두 개의 함수, loss와 optimizer를 지정해야한다.  
* `Loss Function`은 알려진 정답과 비교하여 추측된 정답을 측정하고 정답이 얼마나 잘 또는 얼마나 잘못되었는지를 측정합니다.   
* `Optimizer Function`은 Loss Function이 실행 된 후 이 Loss를 최소로 만드는 방향으로 실행됩니다.  
* `Epochs`를 이용해서 반복수를 지정하여 위의 함수들을 반복해서 실행해 최적의 함수를 찾게됩니다.  
* 일단 처음에는 Loss function은 `Mean Squared Error`를 Optimizer Function은 `Stochastic Gradient Descent`를 사용할 것이다. 현재로서는 수식을 이해하기 보다 일단 이들을 사용한다는 것만 알고 넘어가도록 하자.


```
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

# Providing the Data

아주 간단한 식을 생성하여 (x 6개, y 6개를 생성하자), 'y=2x-1'이라는 식을 구현해보도록 하자.

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)
```

# Training the Neural Network

Neural Network를 훈련시키는 과정에서 **model.fit**을 사용하여 Xs와 Ys의 관계를 학습시킨다.  

```
model.fit(xs, ys, epochs = 500)
```

위를 실행완료되면 Model fitting이 끝난 것이고, 다음으로는 **model.predict**를 사용해서 원하는 x값에 대한 y값이 제대로 출력되는지 확인하도록 한다.

```
print(model.predict([10.0]))
```

정확히 '19'로 떨어지지 않는 것은 NN이 확률로 처리하기 때문이란 것을 기억하자.  
더불어 데이터가 6개밖에 없기 때문에 우리는 확신할 수 없는 것이다. 그래서 결과가 19 근처로 나오는 것이다.  
우리들은 NN을 다룰 때, 이런 패턴이 반복되는 것을 볼 수 있고 거의 항상 확실성이 아닌 확률을 다루게 될 것이다. 특히 분류에 관한 한, 확률에 기초하는 결과가 무엇인지 알아내기 위해 약간의 코딩 작업을 하게 될 것이다.
